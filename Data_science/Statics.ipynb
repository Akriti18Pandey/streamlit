```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#positive relation
Ads=[10,20,30,40,50]
Sale=[15,25,35,50,65]
```

```python
corr=np.corrcoef(Ads,Sale)
```

```python
corr_value=corr[0,1]
```

```python
corr_value
```

```python
#negative relation
Ads=[10,20,30,40,50]
Sale=[5,3,8,6,1]
```

```python
corr=np.corrcoef(Ads,Sale)
```

```python
corr_value=corr[0,1]
```

```python
corr_value
```

```python
df=pd.read_csv("manufacturing_metrics_dataset.csv")  
df
```

```python
#covariance
cov_matrix=df.cov()
cov_matrix
```

```python
cor_matrix=df.corr()
cor_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.heatmap(cor_matrix, annot= True, cmap="plasma")
plt.title("Correlation matrix in manufacturing")
plt.show()
```
