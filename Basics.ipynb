```python
import pandas as pd
```

```python
# data to series form
marks=[85,91,76,76,56]
std_marks=pd.Series(marks)
print(std_marks)
```

```python

marks=pd.Series([85,91,76,76,56],index=["Math","Science","English","Hindi","Geo"])
print(marks)
```

```python
# updating series by index numbers
marks[1]=55
print(marks)
```

```python
# performing some arithmetic functions

upd_marks=pd.Series([85,91,76,76,56])
print(upd_marks + 5)
```

```python
# Identifying missing value in dataset

missing_marks=pd.Series([85,91,76,None,56])
print(missing_marks)
```

```python
# handling missing value
missing_marks.fillna(50,inplace=True)
print(missing_marks)
```

```python
data={
    "Name":["Amit","Ajeet","Raj","Suraj","Saksham"],
    "Marks":[94,None,75,99,83]
}
df=pd.DataFrame(data)
```

```python
#check null values
df.isnull()
```

```python
#remove null value
df.dropna()
```

```python
#fill null by specify value
df.fillna(70)
```

```python
#fill missing value t= with average of other values
df["Marks"].fillna(df["Marks"].mean(),inplace=True)
```

```python
df
```

```python
# grouping data
```

```python
data={
    "Trade":["AIPA","AIPA","COSMO","CSA","COPA","COSMO","CSA"],
    "Students":[20,20,40,35,30,50,10]
}
df=pd.DataFrame(data)
df
```

```python
df.groupby("Trade")["Students"].sum()
```

```python
df.groupby("Trade")["Students"].mean()
```

```python
df.groupby("Trade")["Students"].count()
```

```python
# used for summarizing data
df.groupby("Trade")["Students"].agg(["sum","mean","count"])
```
